```python
import os
import glob
import pandas as pd

RUTA_BASE = "."

archivos_test = glob.glob(os.path.join(RUTA_BASE, "*_test.csv"))

if not archivos_test:
    raise RuntimeError("No se encontraron archivos *_test.csv")

for ruta in archivos_test:
    nombre_archivo = os.path.basename(ruta)

    # Dataset desde el nombre
    dataset = nombre_archivo.split("_tdataset")[0]

    df = pd.read_csv(ruta)

    # La clase está en la última columna
    columna_clase = df.columns[-1]

    conteo = df[columna_clase].value_counts().sort_index()

    print("\n" + "=" * 70)
    print(f"Dataset: {dataset}")
    print(f"Archivo: {nombre_archivo}")
    print("=" * 70)

    for clase, cantidad in conteo.items():
        print(f"{str(clase):<6} {cantidad}")
```
